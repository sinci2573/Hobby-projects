```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("C://Users//sinci//Downloads//concrete.csv")
```

```python
print(df.head())
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='cement', y='strength', alpha=0.5)
plt.title('Scatter Plot: Cement vs. Strength')
plt.xlabel('Cement')
plt.ylabel('Strength')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='water', y='strength', alpha=0.5)
plt.title('Scatter Plot: Water vs. Strength')
plt.xlabel('Water')
plt.ylabel('Strength')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='strength', alpha=0.5)
plt.title('Scatter Plot: Age vs. Strength')
plt.xlabel('Age')
plt.ylabel('Strength')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='ash', y='strength', alpha=0.5)
plt.title('Scatter Plot: Fly Ash vs. Concrete Strength')
plt.xlabel('Fly Ash')
plt.ylabel('Concrete Strength')
plt.show()
```

```python
sns.lmplot(x="ash",y="strength",data=df)
sns.lmplot(x="cement",y="strength",data=df)
```

```python
plt.figure(figsize=(10, 8))
corr_matrix = df.corr()  
sns.heatmap(corr_matrix)
plt.title('Correlation Heatmap')
plt.show()
```

```python
corrmatrix=df.corr()
round(corrmatrix,2)
```
