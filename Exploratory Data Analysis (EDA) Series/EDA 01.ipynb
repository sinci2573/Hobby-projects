```python
numeric_df = df.select_dtypes(include=['number'])
range_values = numeric_df.max() - numeric_df.min()
variance_values = numeric_df.var()
std_dev_values = numeric_df.std()
iqr_values = numeric_df.quantile(0.75) - numeric_df.quantile(0.25)

print("Range:\n", range_values)
print("\nVariance:\n", variance_values)
print("\nStandard Deviation:\n", std_dev_values)
print("\nInterquartile Range (IQR):\n", iqr_values)
```

```python
numeric_df = df.select_dtypes(include=['number'])

mean_values = numeric_df.mean()
median_values = numeric_df.median()
mode_values = numeric_df.mode().iloc[0]

print("Mean:\n", mean_values)
print("\nMedian:\n", median_values)
print("\nMode:\n", mode_values)
```
