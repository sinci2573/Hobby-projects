```python
import pandas as pd
customer_info=pd.read_csv("C://Users//sinci//Downloads//Customer_Info.csv")

order_history=pd.read_csv("C://Users//sinci//Downloads//Order_History.csv")
```

```python
order_history['Order_Date'] = pd.to_datetime(order_history['Order_Date'], errors='coerce', dayfirst=True)
customer_info['Date_of_Birth'] = pd.to_datetime(customer_info['Date_of_Birth'], errors='coerce', dayfirst=True)
```

```python
order_history['Quantity'] = pd.to_numeric(order_history['Quantity'], errors='coerce')
```

```python
merged_data = pd.merge(order_history, customer_info, on='Customer_ID', how='inner')
merged_data
```

```python
merged_data = pd.merge(order_history, customer_info, on='Customer_ID', how='outer')
merged_data
```

```python
merged_data.drop_duplicates(subset=['Customer_ID', 'Email'], keep='first', inplace=True)
merged_data
```

```python
merged_data = pd.merge(order_history, customer_info, on='Customer_ID', how='left')
merged_data
```

```python
merged_data = pd.merge(order_history, customer_info, on='Customer_ID', how='right')
merged_data
```

```python
merged_data.to_csv('Merged_Customer_Order_Data.csv', index=False)
merged_data                  
```
